```python
def convert_loan_type(value):
  if value in ['Personal', 'Personal Loan', 'personal', 'PERSONAL']:
    return 'personal'
  elif value in ['MORTGAGE', 'mortgage', 'Mortgage']:
    return 'mortgage'
  elif value in ['Credit Card', 'CreditCard', 'credit card', 'CC']:
    return 'credit card'
  else:
    return value

df['loan_type'] = df['loan_type'].map(convert_loan_type)


def convert_employment_type(value):
  if value in ['Full-time', 'Full Time', 'FULL_TIME', 'Fulltime', 'FT']:
    return 'Full-time'
  elif value in ['SELF_EMPLOYED', 'Self Emp', 'Self Employed', 'Self-employed']:
    return 'Self-employed'
  elif value in ['PART_TIME', 'Part-time', 'Part Time', 'PT']:
    return 'Part-time'
  elif value in ['Contractor', 'CONTRACT', 'Contract']:
    return 'Contract'

df['employment_type'] = df['employment_type'].apply(convert_employment_type)
```

```python
financial_cols = [
    'credit_usage_amount',
    'available_credit',
    'total_monthly_debt_payment',
    'total_debt_amount',
    'monthly_free_cash_flow',
]

for col in financial_cols:
    df[col] = (
        df[col]
        .astype(str) # Ensure it's treated as a string series first
        .str.replace(r'[\$,]', '', regex=True)
        .astype(float)
    )
```
