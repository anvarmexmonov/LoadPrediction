```python
import pandas as pd
```

```python
demographics_df = pd.read_csv('demographics.csv')
app_meta_df = pd.read_csv('application_metadata.csv')
financial_ratios = pd.read_json("financial_ratios.jsonl", lines=True)
loan_details = pd.read_excel('loan_details.xlsx')
credit_history = pd.read_parquet("credit_history.parquet")
geographic_df = pd.read_xml('geographic_data.xml')
```

```python
merged_df = pd.concat([loan_details, financial_ratios, demographics_df, geographic_df, credit_history, app_meta_df], axis=1)
```

```python
merged_df.isnull().sum()
```

```python
merged_df.to_csv('validation_set.csv', index=False)
```
